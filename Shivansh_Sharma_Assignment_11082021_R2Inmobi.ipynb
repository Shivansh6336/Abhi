```python
#Loading Packages
import pandas as pd
```

```python
int_data = pd.read_csv(r'Interview_Dataset_2807.csv')
print(int_data.shape)
int_data.head(5)
```

```python
#Understanding no. of columns and rows with data types
int_data.info()
```

```python
#Removing null column
int_data = int_data.drop('PMP Deal ID',axis = 1)
```

```python
int_data.head(30)
```

```python
#Loading column names
int_data.columns
```

```python
#Treating Dates
int_data['Week Start Date']= pd.to_datetime(int_data['Week Start Date'],infer_datetime_format= True)
int_data['Week End Date']= pd.to_datetime(int_data['Week End Date'],infer_datetime_format= True)
```

```python
#Treating Dates another way
int_data['Week Start Date']= pd.to_datetime(int_data['Week Start Date'],unit= 's')
#int_data['Week End Date']= pd.to_datetime(int_data['Week End Date'],infer_datetime_format= True)
int_data.head(30)
```

```python
#Replacing special characters with 0
int_data['Burn'] = int_data['Burn'].replace("-",0)
int_data['Burn'] = int_data['Burn'].replace("**",0)
int_data['Burn'] = int_data['Burn'].replace("Not Available",0)
int_data.head(30)
```

```python
int_data['Burn'] = int_data['Burn'].astype(float)
```

```python
#Checking which account has max and min burn
maxmin_burn_acc = int_data.groupby(['Account Name']).agg({'Burn':sum}).reset_index().sort_values(by='Burn')
```

```python
#Checking which account has max and min burn by channel
maxmin_burn_acc_channel = int_data.groupby(['Account Name','Special classification']).agg({'Burn':sum}).reset_index().sort_values(by='Burn')
```
